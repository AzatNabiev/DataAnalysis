#Гистограмма


Гистограмма — способ представления табличных данных в графическом виде — в виде столбчатой диаграммы.

Суть данного метода заключается в том, что мы разбиваем область определения значений нашей выборки на r интервалов, после чего считаем для каждого интервала количество попаданий в него. Далее графически над каждым интервалом надстраиваем столбец, высота которого соответствует количеству принадлежащих ему значений выборки.


Гистограмма используется в качестве наглядного изображения выборочной плотности. В нашей задаче такой постановки задачи не было, поэтому мы использовали её, чтобы увидеть распределение значений (например, количества добавленных или обновлённых записей) на временной прямой.

#Диаграмма

Диаграмма — графическое представление данных линейными отрезками или геометрическими фигурами, позволяющее быстро оценить соотношение нескольких величин.

В нашей задачи анализа мы распространённо использовали данный метод, ведь сырые выводы, сделанные программой, не самым лучшим образом иллюстрируют результат, как для нас самих, так и для конечных обозревателей. Поэтому в нашей задачи мы использовали гистограммы, диаграммы, графики, а также интерактивные карты, на которых размер кружка-города зависел от количества отелей в нём.

#Линейная регрессия

Линейная регрессия — модель зависимости переменной x от одной или нескольких других переменных (факторов, регрессоров, независимых переменных) с линейной функцией зависимости

Суть метода  сводится к тому, что мы получаем и используем уравнение вида:

y = a + b1*x1 + b2*x2... +bn*xn 

И используем его для того, чтобы предсказывать значение y на основе вектора значений факторов. Это одна из простейший моделей предсказания.

Вычислять данные кожффициенты по методу наименьших квадратов: мы хотим, чтобы отклонение предсказанных значений от реальных было минимальным. Составляем функцию - сумму квадратов разностей значений, берём частные производные, решаем систему уравнений. В итоге получаем коэффициенты, зависящие от коэффициентов корреляции - понятное дело, ведь корелляция отображает линейную зависимость между величинами.

#Pandas

Pandas - удобная библиотека для работы с таблицами. В нашем случае она будет незаменимой, так как наш датасет - это одна большая таблица. Данное расширение позволяет загрузить таблицу из файла (excel, csv...), обработать данные (пройтись по пустым строчкам и удалить их, переименовать колонки и многое другое), а после сгруппировать данные или вычислить что-то на их основе (опять же, прямо-таки с помощью функций библиотеки)

#Plotly

Plotly - мощный инструмент визуализации, применяемый не только в питоне. Позволяет в достаточно удобных обозначениях визуализировать графики в виде от простой гистограммы до интерактивной карты или 3D анимации. Вся визуализация в нашем проекте будет построена именно на этой библиотеке

#Numpy

Пакет numpy используется практически во всех численных вычислениях с помощью Python. Это пакет, предоставляющий высокопроизводительные векторные, матричные и более высокоразмерные структуры данных для Python. В сущности, прямых использований данной библиотеки у нас нет (или незначительно мало), но Pandas частично связан с данной библиоткей, + это один из стандартных импортов в проект
